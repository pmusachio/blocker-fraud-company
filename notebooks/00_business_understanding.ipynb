# Blocker Fraud Company - Business Understanding

This notebook documents the business problem, assumptions, success metric and analytical plan.

## Business Problem

A financial operation needs to block fraudulent transactions without slowing down legitimate, high-value transfers.

## Goal

Detect fraudulent financial transactions while prioritizing recall, precision@k and protected financial impact.

## Key metric

Average precision, recall, precision@k and expected protected revenue

## Business Assumptions

- Fraud is rare, so plain accuracy does not represent business quality.
- False negatives cause larger financial loss than the extra review time spent on false positives.
- The model score should feed a prioritized investigation queue.

## Solution Strategy

1. **Step 01. Data Description:** Validate schema, dimensions, missing values, types and granularity.
2. **Step 02. Feature Engineering:** Create variables driven by the problem domain.
3. **Step 03. Data Filtering:** Remove records with no analytical value or that risk leakage.
4. **Step 04. Exploratory Data Analysis:** Validate hypotheses and separate relevant signal from noise.
5. **Step 05. Data Preparation:** Impute, scale and encode variables for modeling.
6. **Step 06. Feature Selection:** Separate IDs, target, input variables and discarded columns.
7. **Step 07. Machine Learning Modelling:** Train a reproducible baseline and evaluate technical metrics.
8. **Step 08. Hyperparameter/Fit Strategy:** Reserve room for tuning, threshold calibration and model comparison.
9. **Step 09. Business Translation:** Convert metrics into decisions, prioritization, risk, revenue or operations.
10. **Step 10. Delivery:** Publish the score through an API, a Streamlit dashboard and a Telegram bot for operational use.
