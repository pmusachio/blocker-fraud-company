# Blocker Fraud Company - Exploratory Analysis

Notebook for validating hypotheses and generating charts for storytelling.

```python
from pathlib import Path
import sys
import json

PROJECT_ROOT = Path.cwd().resolve()
sys.path.insert(0, str(PROJECT_ROOT / "src"))

from blocker_fraud_company.config import load_config
from blocker_fraud_company.data import load_training_frame, profile_dataframe
from blocker_fraud_company.features import prepare_features

config = load_config()
config["project"]
```

```python
df = load_training_frame(config)
df.describe(include='all').T.head(30)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

target = config.get('data', {}).get('target')
if target:
    target = target.lower()
    if target in df.columns:
        ax = df[target].value_counts(normalize=True).head(20).plot(kind='bar', figsize=(10, 4), title=f'Distribuicao de {target}')
        ax.set_ylabel('share')
        plt.show()
```

```python
numeric = df.select_dtypes(include='number')
if not numeric.empty:
    cols = numeric.columns[:8]
    numeric[cols].hist(figsize=(14, 8), bins=30)
    plt.tight_layout()
    plt.show()
```

## Hypotheses under review

- `TRANSFER` and `CASH_OUT` transactions concentrate most of the fraud risk.
- Inconsistent balance deltas are strong fraud signals.
- High amounts must be evaluated together with origin and destination balances.
